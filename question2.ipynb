```python
import numpy as np
import matplotlib.pyplot as plt

# Data
income = [15000,18000,20000,22000,17000,19000,21000,23000,16000,20000,
          18000,22000,21000,24000,17000,19000,20000,23000,18000,21000]

orders = [3,4,5,6,4,5,6,7,3,5,4,6,5,7,4,5,6,7,4,5]

spotify = [1,2,3,4,2,3,4,5,1,3,2,4,3,5,2,3,4,5,2,3]

shop_hours = [2,3,4,5,3,4,5,6,2,4,3,5,4,6,3,4,5,6,3,4]

money_spent = [4000,4500,5000,6000,4200,4800,5500,6500,4100,5000,
               4600,6000,5300,6800,4300,4900,5200,6700,4500,5400]

# Convert to numpy arrays
income = np.array(income)
orders = np.array(orders)
spotify = np.array(spotify)
money_spent = np.array(money_spent)

# Statistics
print("Mean Income:", np.mean(income))
print("Median Income:", np.median(income))
print("Std Deviation Income:", np.std(income))

print("Mean Spending:", np.mean(money_spent))
print("Median Spending:", np.median(money_spent))
print("Std Deviation Spending:", np.std(money_spent))

# Correlation
correlation = np.corrcoef(income, money_spent)[0,1]
print("Correlation (Income vs Spending):", correlation)

# Scatter Plot
plt.scatter(income, money_spent)
plt.title("Income vs Money Spent")
plt.xlabel("Monthly Income")
plt.ylabel("Money Spent")
plt.show()

# Histogram
plt.hist(orders)
plt.title("Distribution of Online Orders")
plt.xlabel("Orders")
plt.ylabel("Frequency")
plt.show()

# Line Plot
plt.plot(spotify)
plt.title("Spotify Listening Hours")
plt.xlabel("User ID")
plt.ylabel("Hours")
plt.show()
```
