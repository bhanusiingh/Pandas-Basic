Group by in Pandas

```python
import pandas as pd
import numpy as np
data ={
    "Name": ["Aman","Saurav","Karan","Shubham","Abhishek",None,"Sameer"],
    "Age": [22,23,None,22,25,24,26],
    "Marks": [85,None, 88,90,56,72,None],
    "Subject": ["Math","Physics","Math","Chemistry","Math","Physics","Chemistry"]
}
df = pd.DataFrame(data)
print(df)
```

```python
#group by subject and calculate mean mark
df.groupby("Subject")["Marks"].mean()
```

```python
df.groupby("Subject")["Name"].count()
```

```python
#Multiple Aggregations
df.groupby('Subject')['Marks'].agg(['mean', 'min', 'max'])
```

```python
df.groupby('Subject')['Marks'].agg(['mean', 'min', 'max','count'])
```

```python
#Group by multiple columns
df.groupby(['Subject','Age'])['Marks'].mean()
```
