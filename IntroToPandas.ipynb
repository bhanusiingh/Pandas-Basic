Importing pandas library

```python
import pandas as pd
import numpy as np
```

Series in pandas

```python
s = pd.Series([10,20,30,40])
print(s)
```

Custom indexing

```python
s= pd.Series([10,20,30], index=["a","b","c"])
print(s)
```

```python
print(s["a"])
```

Series from Dictionary

```python
data= {"Math":90, "Physics":85, "Chemistry":80}
marks = pd.Series(data) 
#to create a Series from a dictionary, where the keys of the dictionary become the index of the Series and the values become the data of the Series.
print(marks)
```

Data Frames

```python
Data= {"Name": ["Jimmy","Karan","Krish"],
       "Age":[14,18,15],
       "Marks":[87,85,90]}
print(Data)
```

```python
type(Data)
```

```python
marks = pd.DataFrame(Data)
print(marks) #Pandas automatically adds a default index.
```

```python
# we can also add custom index in pandas dataframe
marks = pd.DataFrame(Data,index=["Student1","Student2","Student3"])
print(marks)
```

```python
marks.set_index("Name",inplace= True)
print(marks)
```

Handling missing values


```python
#updating the dataset with null values
df=pd.DataFrame({
    "Name":["Aman","saurav","Karan","shubham",None,"Oscar"],
    "Age":[22,23,None,22,25,26],
    "Marks":[85,None,88,90,56,None]
    })
df
```

```python
df.isnull() #to check for null values in the dataset
```

```python
df.isnull().sum() 
#to count the number of null values in each column
```

```python
df.dropna() #to drop rows with null values from the dataset
```
